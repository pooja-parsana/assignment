**Q1. What is Flask Framework? What are the advantages of Flask Framework?**

Flask is a micro web framework for building web applications in Python. It is designed to be lightweight, flexible, and easy to use. Flask is often used for building small to medium-sized web applications that require minimal overhead and quick development time.

Here are some advantages of using Flask Framework:

1. Lightweight and Modular: Flask is designed to be lightweight and modular, making it easy to add or remove components as needed. This also makes it easy to integrate with other tools and frameworks.

2. Simple to Use: Flask has a simple and intuitive syntax that is easy to learn and use. This allows developers to focus on building the application logic rather than worrying about the framework itself.

3. Flexibility: Flask allows developers to build applications in their own way, rather than imposing a specific structure or design pattern. This makes it easy to customize and extend the framework as needed.

4. Easy to Test: Flask applications are easy to test, thanks to its built-in testing support and the ability to run the application in a testing environment.

5. Large Community: Flask has a large and active community of developers who contribute to the development of the framework, as well as provide support and resources for other developers.

6. Scalability: Flask is scalable, meaning that it can handle large amounts of traffic and data with ease. It can be used for building both small and large applications.

In summary, Flask is a lightweight and flexible web framework that is easy to use and customize. Its simple syntax and large community make it a popular choice for building web applications in Python.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing is a mechanism that allows us to map URLs (Uniform Resource Locators) to specific functions in our application. It defines the URL patterns that an application will respond to and what function will be executed when a particular URL is requested.

We use app routing in Flask to handle HTTP requests and provide responses to clients. The `@app.route()` decorator in Flask is used to bind a function to a specific URL. The URL can also have variable parts that can be extracted by the function and used in the response.

For example, consider the following code:

```python
@app.route('/hello')
def hello():
    return 'Hello, World!'
```

In the above code, the `@app.route()` decorator binds the `hello()` function to the `/hello` URL. Whenever a client makes a request to the `/hello` URL, Flask calls the `hello()` function and returns the string "Hello, World!" as the response.

We use app routes in Flask to create a RESTful API, handle user input, and provide dynamic content to users. It allows us to create a URL structure that is easy to understand and navigate for users, and also makes it easier to maintain the application.

Overall, app routing is an essential feature of Flask that allows developers to define the URL structure of an application and handle HTTP requests in a flexible and dynamic manner.

**Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.**

In Flask, the `url_for()` function is used for URL building. This function generates a URL for the specified endpoint and takes the endpoint name as the first argument.

Here's an example code to demonstrate the working of the `url_for()` function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the index page.'

@app.route('/user/<username>')
def profile(username):
    return f'This is the profile page for {username}.'

@app.route('/login')
def login():
    return 'This is the login page.'

with app.test_request_context():
    print(url_for('index'))        
    print(url_for('login'))        
    print(url_for('profile', username='John Doe')) 
```

In the above code, we have defined three routes - `/`, `/login`, and `/user/<username>`. The `url_for()` function is used to generate URLs for these routes.

Inside the `with app.test_request_context()` block, we have called the `url_for()` function for the `index`, `login`, and `profile` endpoints. The output of these calls is printed to the console.

When we call `url_for('index')`, Flask generates the URL `/` for the `index` endpoint. Similarly, `url_for('login')` generates the URL `/login` for the `login` endpoint. When we call `url_for('profile', username='John Doe')`, Flask generates the URL `/user/John%20Doe` for the `profile` endpoint with the `username` variable set to `'John Doe'`.

By using the `url_for()` function, we can avoid hardcoding URLs in our code and ensure that the URLs are generated dynamically based on the endpoint name and any associated variable parts.
