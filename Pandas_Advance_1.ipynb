info(): This function provides a concise summary of a DataFrame, including the column names, data types, and the number of non-null values.


````python
import pandas as pd

# Create a DataFrame
data = {'Name': ['John', 'Alice', 'Bob'],
        'Age': [25, 28, None],
        'City': ['New York', 'London', 'Paris']}

df = pd.DataFrame(data)

# Display the summary information of the DataFrame
df.info()
````

describe(): This function generates descriptive statistics of numerical columns in a DataFrame, such as count, mean, standard deviation, minimum, and maximum values.

````python
import pandas as pd

# Create a DataFrame
data = {'Name': ['John', 'Alice', 'Bob'],
        'Age': [25, 28, 32],
        'Salary': [50000, 60000, 70000]}

df = pd.DataFrame(data)

# Generate descriptive statistics of the DataFrame
print(df.describe())
````

groupby(): This function is used to group rows in a DataFrame based on one or more columns, allowing for aggregation and analysis of data within each group.

````python
import pandas as pd

# Create a DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Alice', 'Bob'],
        'Age': [25, 28, 32, 27, 29],
        'Salary': [50000, 60000, 70000, 55000, 65000]}

df = pd.DataFrame(data)

# Group the DataFrame by 'Name' column and calculate the mean salary for each name
grouped = df.groupby('Name')['Salary'].mean()

# Display the result
print(grouped)
````

**Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.**

````python
import pandas as pd
data={
    "A" : 1,
    "B" : 2,
    "C" : 3
}
df=pd.DataFrame(data.items())
df
````

````python
def reset_df_index(df):
  df.reset_index(drop=True,inplace=True)
  df.index =df.index*2 +1
  return df

reset_df_index(df)
````

**Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.**

````python
import pandas as pd
data={
    "Text" : ["POOJA PARSANA", "PARSANA POOJA", "COMPUTER ENGINEER", "ENGINEER"],
}
df=pd.DataFrame(data)
df
````

````python
def word_count(df):
  df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
  return df

word_count(df)
````

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

print(df.size)  
````

DataFrame.shape: The shape attribute returns a tuple representing the dimensions of the DataFrame. It provides the number of rows and columns in the DataFrame as (num_rows, num_columns).

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

print(df.shape) 
````

**Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,  median, and standard deviation of the values in the 'Values' column.**

````python
import pandas as pd

data = {'Values': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)
````

````python
def calculate_statistics(df):
    column_values = df['Values']
    mean_value = column_values.mean()
    median_value = column_values.median()
    std_deviation = column_values.std()
    
    return mean_value, median_value, std_deviation

calculate_statistics(df)
````

**Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.**

````python
import pandas as pd

# Create a sample DataFrame
data = {'Sales': [10, 12, 15, 13, 16, 18, 20, 21, 19, 17, 14, 11]}
dates = pd.date_range('2023-01-01', periods=12, freq='D')
df = pd.DataFrame(data, index=dates, columns=['Sales'])
df
````

````python
def calculate_moving_average(df):
  df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
  return df

calculate_moving_average(df)
````

**Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.**

````python
import pandas as pd
dates = pd.date_range('2023-06-09', periods=7, freq="D")
df = pd.DataFrame({"Date": dates})
df
````

````python
def add_week_days(df):
  df['Weekday'] = df['Date'].dt.strftime("%A")
  return df

add_week_days(df)
````

**Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.**

````python
import pandas as pd

dates = pd.date_range('2023-01-01', periods=60, freq='D')
df = pd.DataFrame({'Date': dates})
df
````

````python

def select_rows_by_date(df):
  start_date = "2023-01-01"
  end_date = "2023-01-31"
  selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
  return selected_rows

select_rows_by_date(df)
````

**Q13. To use the basic functions of pandas, what is the first and foremost necessary library that needs to be imported?**

The first and foremost library that needs to be imported to use the basic functions of pandas is `pandas` itself. The `pandas` library provides the core functionality for working with data in tabular form using DataFrames and Series.

To import the `pandas` library, you can use the following line of code at the beginning of your Python script or notebook:

```python
import pandas as pd
```

By convention, the `pandas` library is commonly imported with the alias `pd`, which allows you to refer to it using the shorthand `pd` throughout your code.

After importing `pandas`, you can then access its various functions and classes to manipulate and analyze data using DataFrames and Series.
