Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
df['Price'].unique()
```

```python
df['Price'] = df['Price'].apply(lambda x: x.replace('$', '') if '$' in str(x) else str(x))
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
average_prices_by_type = df.groupby('Type')['Price'].mean()
```

```python
plt.figure(figsize=(8, 5))
average_prices_by_type.plot(kind='bar', color=['blue', 'green'])
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.

```python
def clean_installs(installs):
    if isinstance(installs, str):
        if '+' in installs:
            return int(installs.replace(',', '').replace('+', ''))
        else:
            return installs
    return installs

df['Installs'] = df['Installs'].apply(clean_installs)

df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

sorted_data = df.sort_values(by='Installs', ascending=False)

top_10_apps = sorted_data.head(10)[['App', 'Installs']]

print(top_10_apps)
```
