Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
```

```python
plt.hist(df['Price'], bins=20, color= "skyblue", edgecolor= "black")
plt.title("Distributions of flight price")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
price_range = df['Price'].max() - df['Price'].min()
max_price = df['Price'].max()
min_price = df['Price'].min()
print(f"Price Range: {price_range}")
print(f"Minimum Price: {min_price}")
print(f"Maximum Price: {max_price}")
```

Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different
airlines.

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x="Airline", y="Price", data=df, palette="viridis")
plt.xticks(rotation=45)
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how
they may impact your analysis.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Price', data=df, palette='viridis')
plt.title('Boxplot of Flight Prices')
plt.xlabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

yes, there is a outliers in datasets.

Above Upper Whisker: Any data point above the upper whisker can be considered a potential outlier. These values might indicate unusually high prices that are significantly different from the majority of prices. Outliers like these could impact the mean and standard deviation, which are sensitive to extreme values.

Below Lower Whisker: Similarly, any data point below the lower whisker can be considered a potential outlier with unusually low prices.

It's important to evaluate the context of your analysis and the nature of the data. Outliers may arise due to data entry errors, genuine extreme values, or other reasons. Depending on the specific dataset and analysis goals, you may choose to handle outliers by removing them, transforming them, or analyzing them separately.

Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight
Price dataset to identify any trends in flight prices. What features would you analyze to identify these
trends, and what visualizations would you use to present your findings to your team?

Analyzing the Flight Price dataset to identify trends in flight prices involves exploring various features that could influence price fluctuations over time. Here are the features you might consider analyzing:

1. **Date of Journey:** Analyze how flight prices change over different months, seasons, and years. Look for any consistent patterns or trends.

2. **Day of the Week:** Investigate how flight prices vary based on the day of the week. Identify whether certain days tend to have lower or higher prices.

3. **Time of Day:** Examine how flight prices change throughout the day. Early morning and late-night flights might have different pricing patterns.

4. **Total Stops:** Analyze the relationship between the number of stops and flight prices. Direct flights vs. multi-stop flights could exhibit distinct pricing trends.

5. **Source and Destination:** Explore how flight prices differ for various source-destination pairs. Some routes might experience more significant price fluctuations than others.

6. **Duration of Stay:** Investigate whether the length of stay affects flight prices. Longer or shorter trips might correlate with specific pricing patterns.

7. **Airline:** Analyze how different airlines' prices change over time. Some airlines might consistently offer lower prices or have more volatile pricing.

8. **Special Events and Holidays:** Consider holidays, festivals, and major events that could impact flight prices. Analyze how prices fluctuate around these dates.

To present your findings to your team, you can use a combination of visualizations to effectively communicate trends:

1. **Time Series Line Plot:** Create a line plot showing how flight prices change over time (e.g., months or years). Use different lines or colors to represent different airlines or routes for easy comparison.

2. **Heatmap Calendar:** Use a calendar-style heatmap to show how flight prices vary by day of the week and month. This can help identify patterns in pricing based on specific days.

3. **Boxplots:** Present boxplots to show the distribution of flight prices for different categories such as airlines, days of the week, or months. This can highlight the median, quartiles, and potential outliers.

4. **Scatter Plots:** Create scatter plots to visualize the relationship between flight prices and other variables, such as the number of stops or the duration of stay.

5. **Bar Charts:** Use bar charts to compare average flight prices for different airlines, source-destination pairs, or months. Stacked bar charts can also show the composition of price components (e.g., base fare, taxes, fees).

6. **Overlayed Histograms:** Overlay histograms of flight prices for different time periods (e.g., different months or days of the week) to identify patterns and changes in price distribution.

7. **Animated Visualizations:** If possible, create animated visualizations to show how flight prices change over time. This can provide a dynamic view of trends and fluctuations.

8. **Interactive Dashboard:** Consider building an interactive dashboard using tools like Tableau or Power BI. This would allow your team to explore the data and trends themselves.

Ensure that your visualizations are well-labeled, have clear titles and axis labels, and are designed to effectively convey insights. Adding annotations, trends lines, and key observations will make your findings more understandable and actionable for your team.
